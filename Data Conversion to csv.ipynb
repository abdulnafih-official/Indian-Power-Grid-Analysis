```python
import pandas as pd
import os

all_dfs = []

for file in os.listdir('Datasets'):
    df_month = pd.read_excel(f'Datasets/{file}', sheet_name='Sheet1', header=1, engine='calamine')
    all_dfs.append(df_month)
    print(f"✅ {file}")

df = pd.concat(all_dfs, ignore_index=True)
print(df.shape)
print(df.head())

df.to_csv('energy_combined.csv', index=False)
print("Saved!")
```
